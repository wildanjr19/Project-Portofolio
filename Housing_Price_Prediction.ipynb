# Feature Engineering

```python
# Binary Encoding
data_encode = {"mainroad" : {"yes":1,"no":0}}
data = data.replace(data_encode)
data_encode = {"guestroom" : {"yes":1,"no":0}}
data = data.replace(data_encode)
data_encode = {"basement" : {"yes":1,"no":0}}
data = data.replace(data_encode)
data_encode = {"hotwaterheating" : {"yes":1,"no":0}}
data = data.replace(data_encode)
data_encode = {"airconditioning" : {"yes":1,"no":0}}
data = data.replace(data_encode)
data_encode = {"prefarea" : {"yes":1,"no":0}}
data = data.replace(data_encode)
```

```python
data
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
data["furnishingstatus"]=LE.fit_transform(data["furnishingstatus"])
```

```python
data
```

```python
# Cek kembali data
for i in data.columns:
  print(data[i].value_counts())
```

# Modelling

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
x = data.drop("price", axis=1)
y = data["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
DT = DecisionTreeRegressor()
```

```python
DT.fit(x_train, y_train)
```

```python
y_pred = DT.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
# Cek akurasi
print(r2_score(y_pred, y_test))
```
